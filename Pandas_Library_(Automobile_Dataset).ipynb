# **PANDAS (PYTHON LIBRARY)**

We shall now test your skills in using Pandas package. We will be using the [automobiles Dataset](https://www.kaggle.com/nisargpatel/automobiles/data) from Kaggle.

```python
import pandas as pd
```

**Load the Automobile dataset into variable "full_data"**

```python
full_data=pd.read_csv("/content/Automobile.csv")
```

**Check the head of the DataFrame.**

```python
full_data.head()
```

**How many rows and columns are there?**

```python
full_data.shape
```

**What is the average Price of all cars in the dataset?**

```python
full_data['price'].mean()
```

```python
full_data['price'].median()
```

**Which is the cheapest make and costliest make of car in the lot?**

```python
full_data[full_data['price']==full_data['price'].max()]
```

```python
full_data[full_data['price']==full_data['price'].min()]
```

**How many cars have horsepower greater than 100?**

```python
full_data[full_data['horsepower']>100].count()
```

**How many hatchback cars are in the dataset ?**

```python
full_data[full_data['body_style']=='hatchback'].count()
```

**What are the 3 most commonly found cars in the dataset?**

```python
full_data['make'].value_counts().head(3)
```

**Someone purchased a car for 7099, what is the make of the car?**

```python
full_data[full_data['price']==7099]['make']
```

**Which cars are priced greater than 40000?**

```python
full_data[full_data['price']>40000]
```

**Which are the cars that are both a sedan and priced less than 7000?**

```python
full_data[(full_data['body_style']=='sedan') & (full_data['price'] < 7000)]
```
